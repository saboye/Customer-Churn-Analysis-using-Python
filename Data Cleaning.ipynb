## Data Cleaning 

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('raw_data/Customer Portfolio Snapshot.csv')
```

```python
data1.head()
```

```python
data2=pd.read_csv('raw_data/Customer Investment Snapshot.csv')
```

```python
data2.head()
```

```python
data3=pd.read_csv('raw_data/Customer Demographics.csv')
```

```python
data3.head()
```

```python
data4=pd.read_csv('raw_data/Customer Attrition Status.csv')
```

```python
data4.head()
```

```python
df_dummy = pd.merge((pd.merge((pd.merge(data1, data2, on='CustomerId')),data3, on='CustomerId')),data4, on='CustomerId')

df_dummy.head()
```

```python
df=df_dummy.drop(['RowNumber_x', 'RowNumber_y'], axis = 1)
```

```python
df.head()
```

```python
df.duplicated()
```

```python
df.describe()
```

#### Checking for True duplicates in Data

```python
df[df[['CustomerId', 'Surname', 'CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance',
       'NumOfProducts', 'HasChckng', 'IsActiveMember', 'EstimatedSalary', 'Exited']].duplicated() == True]
```

```python
subset = ['CustomerId', 'Surname', 'CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance',
       'NumOfProducts', 'HasChckng', 'IsActiveMember', 'EstimatedSalary', 'Exited']
```

```python
df=df.drop_duplicates(subset,keep = 'last').reset_index(drop = True)
```

```python
df[df[['CustomerId']].duplicated() == True]
```

```python
duplicateDFRow = df[df.duplicated()]
print(duplicateDFRow)
```

```python
df.to_csv('dataset/clean_data.csv', index=False)
```
